# Sacar el vocabulario ddel corpus

```python
# abrimos el archivo usando open
with open("/kaggle/input/01-corpus-turismo-500/01_corpus_turismo_500.txt", "r", encoding="utf-8") as f:
    texto = f.read()

texto
```

```python
vocabulario = list(set(texto.lower().replace(' ','\n').replace('.','').replace(',','').split()))
vocabulario
```

```python
len(vocabulario)
```

vamos a generar el corpus

```python
corpus = texto.split('\n')
corpus
```

```python
def get_tf(doc):
    vec = []
    for word in vocabulario:
        cont = doc.lower().replace('.','').replace(',','').split().count(word)
        vec.append(cont)
    return np.array(vec)    
```
