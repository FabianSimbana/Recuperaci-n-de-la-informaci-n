```python
df = pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')

def buscar(docs, querry, na=False):
    #la funcion busca una palabra especifica dentro del dataset
    mask = docs.str.contains(querry)
    return docs[mask]
```

## Parte 2: Construcción de un índice invertido


```python
def construir_indice_invertido(doc, review):
    # contruye un indice invertido a partir de la coluna review de nuestro dataset y lo almacena en un diccionario
    diccionario_palabras = {}

    for i, fila in doc[review].dropna().head(2000).items():
        palabras = str(fila).split()
        for palabra in palabras:
            palabra = palabra.lower().strip(".,!¡¿?:;\"'()[]") 
            if palabra not in diccionario_palabras:
                diccionario_palabras[palabra] = [i]
            else:
                if i not in diccionario_palabras[palabra]:
                    diccionario_palabras[palabra].append(i)
    
    return diccionario_palabras
```

```python
indice_invertido = construir_indice_invertido(df, 'review')
```

```python
#buscamos los indices del dataset que contengan la palabra good
indice_invertido['good']
```
